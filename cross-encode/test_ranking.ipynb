```python
import json
import pandas as pd
```

```python
# JSON file
f = open ('generated-questions.json', "r")
  
# Reading from file
data = json.loads(f.read())
```

```python
list_all=[]
for key, value in data.items():
    for i in value['questions']:
        list_all.append((value['context'], i, 1))
```

```python
df = pd.DataFrame(list_all, columns = ['premise', 'hypo', 'label'])
```

```python
df.to_csv('nli_dataset.csv', index=False)
```
